# Hyperparameter optimization with grid search

```python
params={"datastore": "False",
        "save_results": "False",
        "id_col": 'compound_id',
        "smiles_col": 'rdkit_smiles',
        "response_cols": 'rank',
        "featurizer": 'ecfp',
        "max_epoch": '20',
        "model_type": 'NN',
        "result_dir": 'drive/MyDrive',
        "dataset_key": 'drive/MyDrive/sample_data_curated.csv'}

drop_out_choice = [[0.2,0.2], [0.3,0.3], [0.4,0.4]]
layer_size_choice = [[500,250], [1000,500], [1500, 750]]
learning_rate_choice = [0.0003, 0.0004, 0.0005]

rfd_list = []
rfs_list = []
rfr_list = []
valid_r2_list = []
for dropouts in drop_out_choice:
    for layer_sizes in layer_size_choice:
        for learning_rate in learning_rate_choice:
            params["dropouts"] = dropouts
            params["layer_sizes"] = layer_sizes
            params["learning_rate"] = learning_rate
            rfd_list.append(dropouts)
            rfs_list.append(layer_sizes)
            rfr_list.append(learning_rate)
            tp = parse.wrapper(params)
            pl = mp.ModelPipeline(tp)
            pl.train_model()
            pred_data = pl.model_wrapper.get_perf_data(subset="valid", epoch_label="best")
            pred_results = pred_data.get_prediction_results()
            valid_r2 = pred_results['r2_score']
            valid_r2_list.append(valid_r2)
            print(f"dropouts: {dropouts}, layer_sizes: {layer_sizes}, learning_rate: {learning_rate}, valid_r2: {valid_r2}")
```

```python
perf_df = pd.DataFrame({"dropouts": rfd_list, "layer_sizes": rfs_list, "learning_rate": rfr_list, "valid_r2_score": valid_r2_list})
perf_df.sort_values(by="valid_r2_score", ascending=False)
```
