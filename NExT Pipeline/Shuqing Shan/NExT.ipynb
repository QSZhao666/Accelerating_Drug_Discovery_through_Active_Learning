## Data analysis

```python
target_name='diversity_3500_scores'
out_name=target_name + '_escape.csv'
```

```python
data = pd.read_csv(os.path.join("drive/MyDrive", 'diversity_3500_scores.txt'),sep=' ',header=None)
```

```python
data.columns=['ID','smiles','score']
data.head()
```

```python
#show number of rows in data frame and number of columns
print(data.shape)

# show column names
display(data.columns)

data[['score']].describe()
```

```python
data['rdkit_smiles'] = data['smiles'].apply(curate_data.base_smiles_from_smiles)
```

```python
# curating data
column = 'score'
tolerance = 5
list_bad_duplicates = 'Yes'
max_std = 0.5
compound_id = 'ID'
smiles_col = 'rdkit_smiles'
curated_df = curate_data.average_and_remove_duplicates(column, tolerance, list_bad_duplicates, 
                            data, max_std, 
                            compound_id=compound_id, 
                            smiles_col=smiles_col)
```

```python
# plot distribution histogram
column = 'score'

# # of 'bins' for the histogram
num_bins = 20

title = target_name

units = 'NM'

filepath = ""

cdata=curated_df

curate_data.summarize_data(column, num_bins, title, units, filepath, cdata)
```

```python
curated_df['rank'] = curated_df['score'].rank(ascending=False)
curated_df=curated_df.rename(columns={'ID':'compound_id'})
curated_df.to_csv(os.path.join("drive/MyDrive", 'diversity_3500_scores_curated.csv'))
```

```python
dp.diversity_plots(dset_key = os.path.join("drive/MyDrive", 'diversity_3500_scores_curated.csv'),
                   datastore = False,
                   response_col = 'score',
                   max_for_mcs = 100)
```

```python
feat_type = 'ECFP'
dist_metric = 'tanimoto'
smiles_lst1 = curated_df['rdkit_smiles'].tolist()
calc_type = 'nearest'
dist_sample = cd.calc_dist_smiles(feat_type, dist_metric, smiles_lst1, None, calc_type)
```

```python
from scipy.stats.kde import gaussian_kde

# import math library
import numpy as np

# for creating plots
import matplotlib.pyplot as plt

# current directory
odir='./'

# name for the task
task_name='within dataset'

# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.gaussian_kde.html
dist_pdf = gaussian_kde(dist_sample)
x_plt = np.linspace(min(dist_sample), max(dist_sample), 500)
y_plt = dist_pdf(x_plt)    
fig, ax = plt.subplots(figsize=(8.0,8.0))
ax.plot(x_plt, y_plt, color='forestgreen')
ax.set_xlabel('%s distance' % dist_metric)
ax.set_ylabel('Density')
ax.set_title("%s dataset\nDistribution of %s distances between %s feature vectors" % (
              task_name, dist_metric, feat_type))
```

```python
from sklearn import metrics
print('r2_score',metrics.r2_score(sorted_result['actual'],sorted_result['pred_rank']))
def active(ls,percentage=0.04,n=0):
     total=len(ls)
     ls=ls.rank(ascending=False)
     if n==0:
        newls=np.where(ls<=total*percentage,1,0)
     else:
        newls=np.where(ls<=n,1,0)
     return newls
y_true=sorted_result['actual']
y_true=active(y_true)
y_pred=sorted_result['pred_rank']
y_pred=active(y_pred)
print("accuracy_score:", metrics.accuracy_score(y_true, y_pred))
print("precision_score:", metrics.precision_score(y_true, y_pred))
print("recall_score:", metrics.recall_score(y_true, y_pred))
print("f1_score:", metrics.f1_score(y_true, y_pred))
print("auc_roc:", metrics.roc_auc_score(y_true, y_pred))
```
