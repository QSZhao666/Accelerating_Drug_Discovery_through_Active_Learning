```python
# separate file into manageable chunk, from previous glance, 2000000 row would be around 100 mb
chunksize = 2000000
filename = "d4_zinc_ligs.csv"
part = 0
with pd.read_csv(filename, chunksize=chunksize) as reader:
    for chunk in reader:
        part += 1
        split = pd.DataFrame(chunk)
        outname = "ligs_split_" + str(part) + ".csv"
        split.to_csv(outname, index=False)
        print(outname)
```

```python
filename = "d4_full_best_score_sorted.csv"
chunksize = 2000000
part = 0
with pd.read_csv(filename, chunksize=chunksize) as reader:
    for chunk in reader:
        part += 1
        split = pd.DataFrame(chunk)
        outname = "score_split_" + str(part) + ".csv"
        split.to_csv(outname, index=False)
        print(outname)
```

```python
# join each chunk of the first file with every other chunk of the other file
parts = part + 1
out = 0
for ligs_part in tqdm(range(1,parts)):
    ligsname = "ligs_split_" + str(ligs_part) + ".csv"
    ligs = pd.read_csv(ligsname)
    for score_part in tqdm(range(1,parts)):
        scorename = "score_split_" + str(score_part) + ".csv"
        score = pd.read_csv(scorename)
        score = score.rename(columns={'title':'ID'})
        merged = pd.merge(ligs, score, on='ID')
        out += 1
        outname = "merged_" + str(out) + ".csv"
        merged.to_csv(outname, index=False)
        
```
